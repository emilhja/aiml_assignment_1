## What Each Test Tries To Measure

This section explains the purpose of each tuning run before comparing the metrics.

```python
descriptions = [
    {
        'run': spec['run_name'],
        'description': spec.get('description', ''),
    }
    for spec in RUN_SPECS
]
pd.DataFrame(descriptions)
```

## Summary Table

```python
rows = []
for run_name in RUN_NAMES:
    run_dir = Path(RUN_DIRS[run_name])
    config = load_json(run_dir / 'config.json')
    summary = load_json(run_dir / 'summary.json')
    spec = next(spec for spec in RUN_SPECS if spec['run_name'] == run_name)
    rows.append({
        'run': run_name,
        'description': spec.get('description', ''),
        'model': config['model_name'],
        'conv_channels': config.get('conv_channels'),
        'activation': config.get('activation'),
        'dropout': config.get('dropout', 0.0),
        'batch_norm': config.get('batch_norm', False),
        'weight_decay': config.get('weight_decay', 0.0),
        'l1_lambda': config.get('l1_lambda', 0.0),
        'input_noise_std': config.get('input_noise_std', 0.0),
        'learning_rate': config.get('learning_rate'),
        'adam_beta1': config.get('adam_beta1'),
        'adam_beta2': config.get('adam_beta2'),
        'adam_eps': config.get('adam_eps'),
        'augmentation_enabled': config.get('augmentation_enabled'),
        'test_acc': summary['final_test_accuracy'],
        'best_val_acc': summary['best_validation_accuracy'],
        'best_val_loss': summary['best_validation_loss'],
        'time_s': summary['total_training_time_seconds'],
        'params': config.get('trainable_parameters', 0),
    })
sorted(rows, key=lambda row: row['test_acc'], reverse=True)
```

## Pandas Overview

```python
df = pd.DataFrame(rows)
display_columns = [
    'run',
    'description',
    'model',
    'conv_channels',
    'activation',
    'dropout',
    'batch_norm',
    'weight_decay',
    'l1_lambda',
    'input_noise_std',
    'learning_rate',
    'adam_beta1',
    'adam_beta2',
    'adam_eps',
    'augmentation_enabled',
    'params',
    'best_val_loss',
    'best_val_acc',
    'test_acc',
    'time_s',
]
df[display_columns].sort_values(by='test_acc', ascending=False).reset_index(drop=True)
```

## Tuning Takeaway

Hyperparameter tuning means trying many plausible settings, then selecting the best trade-off between validation behavior, test accuracy, runtime, and model size.

```python
best_by_test = max(rows, key=lambda row: row['test_acc'])
best_by_val_loss = min(rows, key=lambda row: row['best_val_loss'])
{'best_by_test': best_by_test, 'best_by_val_loss': best_by_val_loss}
```

```python
best_test = max(rows, key=lambda row: row['test_acc'])
best_val_acc = max(rows, key=lambda row: row['best_val_acc'])
best_val_loss = min(rows, key=lambda row: row['best_val_loss'])
fastest = min(rows, key=lambda row: row['time_s'])
smallest = min((row for row in rows if row['params'] > 0), key=lambda row: row['params'])

lines = [
    'English summary:',
    (
        f"Best test run: {best_test['run']} ({best_test['test_acc']:.2%}) using {best_test['model']}. "
        f"Best validation accuracy: {best_val_acc['run']} ({best_val_acc['best_val_acc']:.2%}). "
        f"Best validation loss: {best_val_loss['run']} ({best_val_loss['best_val_loss']:.4f})."
    ),
    (
        f"Fastest run: {fastest['run']} ({fastest['time_s']:.2f}s). "
        f"Smallest model: {smallest['run']} ({smallest['params']:,} trainable parameters)."
    ),
]

print('\n'.join(lines))
lines
```
